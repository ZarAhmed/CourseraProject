Is there a significant difference in median value of houses bounded by the Charles river or not? (T-test for independent samples)

Null Hypothesis : The average are same
Alternate Hypothesis : The average are not same

```python
a = df[df['CHAS'] == 0]['MEDV']
a
```

```python
b = df[df['CHAS'] == 1]['MEDV']
b
```

```python
scipy.stats.ttest_ind(a,b, axis=0, equal_var= True)
```

Since the p value is greater than 0.05, we failed to reject null hypothesis

Is there a difference in Median values of houses (MEDV) for each proportion of owner occupied units built prior to 1940 (AGE)? (ANOVA)


```python
df['AGE'].value_counts()
```

```python
df['age group']
```

Null Hypothesis : All the population means are equal
Alternate Hypothesis : Atleast one differ

```python
low = df[df['age group'] == '35 years and younger']['MEDV']
mid = df[df["age group"] == 'between 35 and 70 years']['MEDV']
high = df[df['age group'] == '70 years and older']['MEDV']
```

```python
f_value, p_value = scipy.stats.f_oneway(low,mid,high, axis = 0)
f_value, p_value
```

Since p value is greater than 0.05, we fail to reject null hypothesis.


Can we conclude that there is no relationship between Nitric oxide concentrations and proportion of non-retail business acres per town? (Pearson Correlation)

Null Hypothesis : They are correlated
Alternate Hyothesis : They are not correlated

```python
pearson, p_value = scipy.stats.pearsonr(df['NOX'],df['INDUS'])
print("Pearson Coefficient value={0}, P-value={0}".format(pearson,p_value))
```

We fail to reject null hypothesis
